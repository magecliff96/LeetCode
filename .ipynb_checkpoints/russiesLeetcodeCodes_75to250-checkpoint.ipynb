# Problem 83: Remove Duplicates from Sorted List (completed) x

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        if head==None or head.next==None:
            return head
        
        p1=head;p2=head.next
        while True:
            while p1.val==p2.val:
                p2=p2.next
                p1.next=p2
                if p2==None:
                    return head
            
            p1=p1.next
            p2=p2.next
            if p2==None:
                return head
            
```

# Problem 94: Binary Tree Inorder Traversal (completed) x

```python
#we need to make sure we need to be in this order: left, root, right
#hence every if cases needs to be in left root right order
#can also use loop + stack to impliment this

#this can be resolved by recongizing the number of possible cases
#technically there are 4 cases, which should be assigned in this order:
#1- the node is a node with no child (just append)
#2- the node is a left node with only a right child (append first then go to right child)
#3- the node is a right node with only a left child (go to left child first then append)
#4- the node is a root node (go to left child first, then append, then right child)
#since if following nodes have no child, it will always lead to none, as long as we resolve the 0 None case,
# we will only need case 1 and 4

#we have 3 functions: 
#append 
#go to a left child 
#go to right child
#always in: Left , append ,right if possible


## first try algorithm
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]: 
        
        
        def recursion(root,ans):#we must make sure append only happens once
            
            if root==None: #if root is empty case
                return
            if root.left == None and root.right != None: ##left child with only right (unneccessary)
                ans.append(root.val) 
                recursion(root.right,ans) 
            elif root.right == None and root.left == None: ##no child
                ans.append(root.val) 
            else: ## have both child (centers) #root/right child with only left/left with only right
                recursion(root.left,ans) #left
                ans.append(root.val) #root
                recursion(root.right,ans) #right, if only left this will lead to none node
        
        ans=[]
        recursion(root,ans)
        return ans

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        def recursion(root,ans):
            if root==None:
                return
            if root.right == None and root.left == None:
                ans.append(root.val)
            else:
                recursion(root.left,ans)
                ans.append(root.val)
                recursion(root.right,ans)
        
        ans=[]
        recursion(root,ans)
        return ans
```

# Problem 101: Symmetric Tree (completed) x

```python
#literally exact same question as problem 100 but must reverse the sides when traversing the tree
#since its a reflection

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        r=root.right
        l=root.left
        
        def recursion(p,q,ans):
            if p!=None and q !=None:  #not None cases
                if p.val==q.val: #equal (proceed)
                    recursion(p.right,q.left,ans) #since its a reflection, reverse progression
                    recursion(p.left,q.right,ans)
                    return 
                else: #unequal
                    ans.append(False)
                    return
            else: #none cases
                if p==q: #equal end node
                    return
                else: #unequal (false)
                    ans.append(False)
                    return
            
        ans=[]
        recursion(l,r,ans) #proccesses ans
        if len(ans)==0:
            return True
        else:
            return False
```

# Problem 104: Maximum Depth of Binary Tree (completed) x

```python
#we want to transport individual i down the list and add one for every travel
#since we stop at a None node (which is a ndoe that doesnt exist), we will have to -1 by default
#normally i should start at 1 since we only +1 when moving downwards, but accounting for the previous logic
#we can simply start at 0 
#this also help accurately ending the recursion at 0 if root is a None itself

#rest is simply going down the list and finding the end node as we add 1 to i 
#then at the end node compare i with recorded answer to see if it is bigger
#record it if it is

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        
        def recursion(root,i,ans):
            if root==None: #None cases=> check to update answer
                if i>ans: #if larger, give i instead of ans for ans update
                    return i #can also be ans=i then return ans #MIN 
                else:
                    return ans #return ans if ans already bigger
            else:
                ans=recursion(root.right,i+1,ans) #since we are not using deep copy but int we have to
                ans=recursion(root.left,i+1,ans) # do "ans=recursion" rather than just "recursion"
                return ans #always return ans
            
            
        i=0 #start at 0
        ans=0 #start at 0 to account for blank case
        ans=recursion(root,i,ans) 
        return ans
```

# Problem 118: Pascal's Triangle (completed) rrr

```python
#the idea was to set the edges as 1 and then have the center be the sum of 2 values from previous layer

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        ans=[[1]] #initialization
        if numRows == 1:
            return ans
        ans.append([1,1])
        if numRows == 2:
            return ans
        
        #when rows > 3 , we need to build the center
        for i in range(numRows-2): #since we already done the first 2 rows, it has to be row-2
            iAns=ans[-1] #reseting the parent layer to be previous layer 
            temp=[] #empty out temp
            for j in range(len(iAns)+1): #since the next layer will be one larger than parent, +1
                if j == 0 or j == len(ans):#edges
                    temp.append(1) ## building the edges for the next layer
                else: #not edges
                    temp.append(iAns[j]+iAns[j-1]) #buidling center for next layer
            ans.append(temp) #append new after finishing each temp layer
        return ans
```

# Problem 146: LRU Cache (completed) x

```python
class ListNode:
    def __init__(self,val=None, key=None,nextt=None, prev=None):
        self.val=val
        self.key=key
        self.next=nextt
        self.prev=prev

class LinkedList:
    def __init__(self):
        self.head=None #setting to none instead of dummy node to ensure dummy node wont affect capacity
        self.tail=None
    
    def add(self, val,key):
        if self.tail==None: #first node?
            self.tail=ListNode(val,key) #intialize first node
            self.head=self.tail #create head from tail
        else:
            self.head.prev=ListNode(val,key,self.head) #intialize node
            self.head=self.head.prev #update head to frontest node
            
    def removeHead(self): 
        if self.head!=self.tail:#Only valid when tail!=head
            temp=self.head
            self.head=self.head.next
            temp.next=None
            self.head.prev=None
            return temp.key
        else: #if deleting only node left, reset to empty
            temp=self.head #to return key
            self.head=None
            self.tail=None
            return temp.key
        
    def removeTail(self): 
        if self.head!=self.tail:#Only valid when tail!=head
            temp=self.tail
            self.tail=self.tail.prev
            temp.prev=None
            self.tail.next=None
            return temp.key
        else: #if deleting only node left, reset to empty
            temp=self.tail #to return key
            self.head=None
            self.tail=None
            return temp.key

    def removeNode(self,node):
        if node==self.head:
            self.removeHead()
        elif node==self.tail:
            self.removeTail()
        else:
            a=node.next
            b=node.prev
            node.next=None;node.prev=None
            a.prev=b
            b.next=a
            
class LRUCache:
    def __init__(self, capacity: int):
        self.keyMapNode={}
        self.capacity=capacity
        self.a=LinkedList()
        
    def get(self, key: int) -> int:
        if key not in self.keyMapNode:
            return -1
        else:
            value=self.keyMapNode[key].val
            self.a.removeNode(self.keyMapNode[key])
            self.a.add(value,key)
            self.keyMapNode[key]=self.a.head
            return value
        
    
    def put(self, key: int, value: int) -> None:

        if key in self.keyMapNode: #update and move node front
            self.a.removeNode(self.keyMapNode[key]) #remove node in middle
            self.a.add(value,key) #readd node in middle
            self.keyMapNode[key]=self.a.head #update node address in dict
        else: #adding node
            if len(self.keyMapNode)==self.capacity: #hitting capacity
                self.a.add(value,key) #adding a new node
                self.keyMapNode[key]=self.a.head #adding node to dictionary
                removedKey=self.a.removeTail() #removing old last node
                self.keyMapNode.pop(removedKey) #removing last node to from dictionary
            else:
                self.a.add(value,key)
                self.keyMapNode[key]=self.a.head
                
            

# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```

# Problem 160: Intersection of Two Linked Lists (completed) rrr

```python
#the idea is to have two pointers on each tail run through the tails first to find the total 
#amount of iteration it takes for A and B to reach the end
#then we compare A and B and see which one is longer
#then we reset the pointers, pA and pB, back to its start, with the longer one getting a head start
#based on the difference in length for list A and list B
#then we will have them both run at same speed. 
#if they run into each other, thats the intersection
#if they dont, that means there is no intersection

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        pA=headA;pB=headB
        iA=0;iB=0 #initialize
        while pA!=None: #counting len of link A (iA)
            pA=pA.next
            iA+=1
        while pB!=None:#counting len of link B (iB)
            pB=pB.next
            iB+=1
        
        pA=headA;pB=headB #reset
        if iA>iB: #if A longer than B, head start for A
            for i in range(iA-iB):
                pA=pA.next
        elif iB>iA: #if B longer than A, head s tart for B
            for i in range(iB-iA):
                pB=pB.next
        
        #now syncronize speed and check if they ever run into each other
        while pA and pB != None:
            if pA==pB: #if they do, thats the intersection 
                return pA
            else: #else forward and stop when reach None
                pA=pA.next
                pB=pB.next
        return None #if reached None and still not intersect, they dont intersect
```

# Problem 167: Two Sum II - Input array is sorted (completed) x

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        s_a=0
        a=0;b=len(nums)-1
        while s_a==0:
            x=nums[a]
            y=nums[b]
            if x+y> target:
                b-=1
            elif x+y< target:
                a+=1
            else:
                s_a==1
                return [a+1,b+1]
```

# Problem 200: Number of Islands (completed) x

```python
## we create a function that will traverse to every point attached to the beginning point
# in the proccess marking the points 0
# we will loop through the entire grid to check every point, running the function on every point to
# disable all points attached to it
# this results in a grid of only seed points. Counting the seed points will result in # islands
# we must becareful with the order of grid[i,j]=0 as that can result in endless recursion
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        
        def recur(grid,i,j,vert,hori):
            grid[i][j]='0' #marking visited
            #Top (-)
            if i>0 and grid[i-1][j]=='1':
                recur(grid,i-1,j,vert,hori)
                
            #Right (+)
            if j<hori-1 and grid[i][j+1]=='1':
                recur(grid,i,j+1,vert,hori)
                
            #Bottom (+)
            if i<vert-1 and grid[i+1][j]=='1':
                recur(grid,i+1,j,vert,hori)
                
            #Left (-)
            if j>0 and grid[i][j-1]=='1':
                recur(grid,i,j-1,vert,hori)
            return
        
        
        islands=0;
        hori=len(grid[0]);
        vert=len(grid);
        
        for i in range(vert):
            for j in range(hori):
                if grid[i][j]=='1':
                    islands+=1
                    recur(grid,i,j,vert,hori)
        return islands
```

```python
#iterative stack reversing val method (no reversing connections)

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        stack=[]; a=head
        while a!=None: #creating stack (since whole system relies on stack, null cases = no stack)
            stack.append(a.val) #no stack = accounted for
            a=a.next
            
        a=head #reset the pointer
        stackLen=len(stack)-1 #MIN
        for i in range(stackLen+1): #reassigning stacks onto the value
            a.val=stack[stackLen-i]        
            a=a.next
        #can also just directly reverse the whole link but since head is attached to the first node,
        #reversing the connection will lead to head.next= None (end of link)
        return head


#stack iterative reversing connection method
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        stack=[]; a=head
        while a!=None: 
            stack.append(a)
            a=a.next
            
        stackLen=len(stack)-1 
        for i in range(stackLen): 
            stack[stackLen-i].next=stack[stackLen-i-1]        

        return stack[stackLen]
#need less to say, stack will take n space

#iterative non stack method
#non stack method is simply having 2-3 pointers and reversing the order
#3 pointer is needed here as we are reversing as we first traverse through the linked list
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        #if head==None:
            #return head (null not needed because we checked whether curr=None)
        
        prev=None
        curr=head
        while curr!=None:
            nextt=curr.next #sets up the next node to ensure forward movement
            curr.next=prev #reroute connection
            
            #moving the pts forward without.next
            prev=curr
            curr=nextt
            
        return prev


#recursive non stack method
#the method first dives deep into the linked list marking each node as curr and prev before resurfacing
#to reverse the connection, resulting in only needing 2 pointers but O(n) space since each
#layer of recursion requires memory
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
    
        def recursion(curr,prev):
            if curr==None: #stopping if end of link, also accounts for Null scenarios
                return prev #this ensure prev as head as we are passing back only the head
            
            #first we dive deep, setting curr to be prev and curr.next will be curr for each layer
            head=recursion(curr.next,curr) 
            curr.next=prev #then we just direct edit, deep copy keeps the edits
            return head 
        
        return recursion(head,None)
```

# Problem 234: Palindrome Linked List (completed) rrr

```python
## idea is we reverse the direction of the link for half of the linked list and 
#then compared the two seperate linked list
#i done this by counting the len, and then restart all the pointers and reverse  the first half
#then compare from the middle
#if it is odd, p2 skips one node (since that would be the center node)
#the p2 list will always be equal or longer than p1 list (since reverse proccess stopps at len//2)

class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        p1=head;p2=head.next;#initualize
        if p2==None: #null cases
            return True
        elif p2.next==None: #null cases so no error when using 3 pters
            if p2.val==p1.val:
                return True
            else:
                return False
        

        ListLen=1;
        while p1.next!=None: #counting len
            p1=p1.next
            ListLen+=1
        s_a=ListLen%2 #check whether its odd or even
        
        #real start
        i1=0 #intitualize and reset
        p1=head;p2=head.next;p3=head.next.next #3 poiinters or 2 pointers with 1 temp pter is needed 
        while i1<ListLen//2:
            if i1!=0: #reversal proccess
                p2.next=p1
                p1=p2
                p2=p3
                p3=p3.next
            else: #initial move (need to reverse head node.next pointing to nein)
                p1.next=None
            i1+=1

        
        if s_a==1: #skipp node if odd
            p2=p2.next #if odd p2 skip
        while p1!=None and p2!=None: #now we have p1 and p2 head of two linked list
            if p1.val!=p2.val: #check if same
                return False #if not gg
            p1=p1.next;p2=p2.next 
        return True #if it passes the check, it means its the same
```

# Problem 236: Lowest Common Ancestor of a Binary Tree (completed) rrr

```python
#the idea was to do recursion and dive and find the node where p and q is
#once encounter p and q, return a corresponding marker as True (a and b respectively) and pass it back
#then after the recursions, we check if a and b is true
#if it is the first time we reach a node where both a and b is true, it means it is the common ancestor
#we save that node and return it upwards along with the true and false 
#then we simply return the node
#this can account for cases even if p or q doesnt exist in the tree
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        def recursion(root,p,q):
            a=False;b=False;ans=None
            if root==p:
                a=True
            if root==q:
                b=True
                
            if root.right!=None:
                [tempA,tempB,tempAns]=recursion(root.right,p,q)
                if tempA:
                    a=True
                if tempB:
                    b=True
                if tempAns!=None:
                    ans=tempAns
            if root.left!=None:
                [tempA,tempB,tempAns]=recursion(root.left,p,q)
                if tempA:
                    a=True
                if tempB:
                    b=True
                if tempAns!=None:
                    ans=tempAns

            if a and b:
                if ans == None:
                    ans=root
                return a,b,ans
            return a,b,ans
            
    
        [a,b,ans]=recursion(root,p,q)
        return ans

#my friend's solution:
#the idea is that instead of using true or false, we will simply pass the node back up if we already]
#found the node
#first we will dive deep and return none if we reached none, if we encountered a p or q, we will 
#return that node
#if we recieved a node instead of a none from prev layer, then
#check if current node is q or p, 
#then return current node if true 
### (this includes cases where say cur is q and p is its child and cur is ancestor)
#or return the node we recieved if false 
### (this means we are just a pass-by node between the q/p and its ancestor)
#if we recieved both node q and p, then it means we are the common ancesstor, then we return current node
# all in all we have 5 cases:
# we are an ancestor recieving both q and p
# we are an ancestor who is a p or q themeslves
# we are p or q 
# we are just a pass-by node
# we are an end node


def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if root == None: #we are end node
            return None #return none
        left = self.lowestCommonAncestor(root.left,p,q) #check if left is relevant
        right = self.lowestCommonAncestor(root.right,p,q) #check if right is relevant
            # below handles 3 cases:
            # we are an ancestor recieving both q and p
            # we are an ancestor who is a p or q themeslves
            # we are p or q
        if root == p or root == q or left == p and right == q or left == q and right == p:
            return root 
        return left if left != None else right #last case: we are just a pass-by node
```

# Problem 240: Search a 2D Matrix II (complete) rrr

```python
#inital idea was to look at the top left corner
## we can tell by the number of first row and first collumn that 
### say example 1, we are looking for 5
### we will search the first row for item bigger than 7
### then we can tell it 5 will be on the collumns on the left of 7
#### we can apply the same thing for the rows, stop at 10
#### we can also do a reversal on the opposite bottom and right rows, but looking for 
#### numbers that are smaller than target this time to limit instead of bigger
####  however, with that , we cant guarantee that its less than mxn


### however, another way to look at it is by looking at the top right corner
### we will soon notice a trend: the item on the left of the current spot is always smaller
### the item on the bottom of the current spot is always bigger...
### basing on that, we can tell that if we look at it from the top right corner, it 
### resembles a binary search tree
### using binary search tree logic: if its smaller we go to the left, if its bigger we go to bottom
### we can easily check whether an item is in the matrix with m+n time
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        if target < matrix[0][0] or target > matrix[-1][-1]:
            return False
        p1=0;p2=len(matrix[0])-1;
        while True:
            pm=matrix[p1][p2] #MIN
            if target==pm: #if target is here, welp return true we found it
                return True
            elif target>pm: #if target is bigger, move to the bottom 
                p1+=1
            else: #if target smaller move to the left
                p2-=1
            
            if p1>=len(matrix): #if either coordinate went out of bound: not in the matrix
                return False
            if p2<0:
                return False
```
