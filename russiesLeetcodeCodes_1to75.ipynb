```python
class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        diff = float('inf')
        nums.sort()
        for i in range(len(nums)):
            lo, hi = i + 1, len(nums) - 1
            while (lo < hi):
                sum = nums[i] + nums[lo] + nums[hi]
                if abs(target - sum) < abs(diff):
                    diff = target - sum
                if sum < target:
                    lo += 1
                else:
                    hi -= 1
            if diff == 0:
                break
        return target - diff
```

```python
#second solution i made much later
#since concat (+) always takes O(n+m), to achieve faster spd, keep temp in array form and only convert 
#to str when appending to ans since append is constant time
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        
        def recursion(digits,mapping,temp,ans):
            if len(digits)==0: #when we reach the end of digits, we append
                ans.append(temp)
                return
            md0=mapping[digits[0]] #MIN
            for i in range(len(md0)): 
                temp+=md0[i] #adding digit and pass down to next layer
                recursion(digits[1:],mapping,temp,ans) #moving onto next digit
                temp=temp[:-1] #resetting the temp we added so next digit we start clean state inherited from prev

        
        mapping={'2':'abc', '3':'def', '4':'ghi', '5':'jkl', '6':'mno', '7':'pqrs', '8':'tuv', '9':'wxyz'} #mapping
        temp='';ans=[]
        if len(digits)==0: #null cases (or else it would return '')
            return ans
        recursion(digits,mapping,temp,ans)
        return ans
```

```python
class Solution: 
    def isValid(self, s: str) -> bool:
        op=['(','[','{']
        dictionary={'(':')','[':']','{':'}'}

        order=[]
        if len(s)<2:
            return False
        for i in range(len(s)):
            temp=s[i]
            if temp in op:
                order.append(temp)
            elif len(order)==0:
                return False
            elif temp==dictionary[order[-1]]:
                order.pop(-1)
            else:
                return False

        if len(order)==0:
            return True
        else:
            return False
```

```python
#Prompt:
#Implement next permutation, which rearranges numbers into the lexicographically next greater permutation of numbers.
#If such an arrangement is not possible, it must rearrange it as the lowest possible order 
#(i.e., sorted in ascending order).
#The replacement must be in place and use only constant extra memory.

#what is lexicographically? confusing right?

#the prompt is very confusing but the goal here is to increase the number formed by the list by as little as u can
#unless u cant, which case u will do a complete miminization/reversion
#by swaping numbers in list(took me 3 hr into coding this to understand the real goal after seeing a post for explaination...)
nums=[1,5,1]
ans=[5,1,1]

def nextPermutation(nums):  
    
    if len(nums)<2: #invalid cases
        return nums

#the biggest number fromed by int in the list is when all the orders are sorted decending order left to right... 
#we will look from the smollest digit to largest to see whether sequence is sorted that way. 
#If yes, we have to reverse entire thing since we cant increase it
#if not, that means we can do some swap and bigger it a little bit

#then first we must find the number we need to swap: any number that is out of order when we see to sort it from big to smol
#in this case it will be the smallest out of order digit we encounter since we also need to minimize the diff post change
#if we pick a bigger out of order digit, we will either not min dif change or just make num smoller 

#then we need to figure out which number to swap to. We have a few options
#bigger nums in higher digits: not possible because if digits ahead are bigger then we just made the num smaller by swap
#smol nums in higher digits: it makes num bigger but there is a better choice than this since we wnana min change
#smol nums in the past: nope, makes num smoller
#bigger nums in the past: yes but which one to min impact?

##to min impact to slot selected digit, find in the past that is slightly bigger (but not smoller as that make num smol)
##that means we reverse check back to see which num that is and do the swap

##once we swap, that means the num is not bigger and nums before swapped digit will be in big to smoll order
##since new digit already bigger, whatever we do in the digits before the swapped digit would not make the num smaller than before
##this way we can just make it in ascending order, minimizing num
#then we done :D

    upperlimit=len(nums)-1 #MIN runtime
    i1=upperlimit;ind=None
    while upperlimit>=i1>0: #loop through the list from the end to start
        i2=i1-1 #initialize left pointer 2  #MIN
        if ind!=None: ##see line 49 first 
            if nums[ind]>=nums[i1]: #see if pointer on right is smoller, we know left things big, left pointer is target 2 if yes
                nums[ind],nums[i2]=nums[i2],nums[ind]
                break 
            elif i1==upperlimit: #everything in the smol digits are bigger, we will take the least biggest one (hence last one) 
                nums[ind],nums[i1]=nums[i1],nums[ind]
                break
            i1+=1 #just counters for progression moving to the right after known target higher digit
        elif nums[i2]<nums[i1]: #comparing if next pointer is targeted higher digit: first one out of decending order L2R
            ind=i2
        else:
            i1-=1 #before we know target higher digit

    if ind==None: #we never found a target higher digit, which means everything is decending order
        ind=0 #starting pt to 0 to just reverse the whole thing
    else:
        ind+=1 #just some calibration to select the right starting pt if THD is found
        
    #since nums before the ind (candidate for swap) is already from largest to smollest, 
    #we reverse that chunk to min number after swap to min num
    nums=reverse(nums,upperlimit,ind) 
    return nums 

def reverse(nums,end,start):
    i=0
    mid=(end-start)/2
    while i<=mid:
        p1=end-i;p2=start+i
        nums[p1],nums[p2]=nums[p2],nums[p1]
        i+=1
    return nums

print(nextPermutation(nums))
```

```python
#the problem is that the array is pivioted, it was sliced at one spot and restacked
#ordinary binary search wont work, but we can break down the secenarios and make binary search work with some finetuning

nums = [4,5,6,7,0,1,2]; target = 4
ans=4
def search(nums, target):
        i=0;j=len(nums)-1; 
        while True:
            k=int((i+j)/2) #index updates
            ni=nums[i];nj=nums[j];nk=nums[k] #MIN
            
#ans check
            if ni==target:
                return i
            elif nj==target:
                return j
            elif nk==target:
                return k
            elif i==k or k==j:
                return -1
            
#two scenario: it is all sorted in search window or still pivioted
            if ni>nj: #pivioted scenario
                if nk>nj: # scenario where mid pt is bigger than end pt, which means right is pivioted and left sorted
                    if ni<target<nk: #scenario where target lays in sorted realm (when it is between num[i] and midpt)
                        j=k-1
                    else: #scenario where target < num[i] or target > num[k] (midpt)
                        i=k+1
                else:# scenario where mid pt is smaller than end pt, which means left is pivioted and right sorted
                    if nj>target>nk: #scenario where target lays in sorted realm (when it is between num[j] and midpt)
                        i=k+1
                    else: #scenario where target > nj or target < nk (midpt) in piviot realm
                        j=k-1
            else: #all sorted scenario (reg binary sort here)
                if nk>target:
                    j=k-1
                else:
                    i=k+1
print(search(nums,target))
```

```python
## the original solution was to try every possible combination and then reset when the combinations reach dead end
# this means keeping a hori vert and square log and writing a recursion that tries out
# every possible combination at each box(node) starting from top to bottom, left to right

## my idea was to replicate the proccess of human solving a sudoku instead
## when a human solves a sudoku, they does not try out every combo from top to bottom
#instead they would look for boxes where there is a certain answer 
#(boxes where there are only 1 answer for sure based on logs)
# then they will fill them in and repeat until they can't progress unless they have to make a choice
#(usually choice of 2 answer and only 1 will lead to the right answer)
# then they will try out one solution, check if it works,
# then repeat above until correct or until they find a empty box where none can fit
# if found a box none can fit, then they back track to the box where they made the last choice
# and so on

# this will help avoid any kind of pointless recursions that are testing out pointless solutions
# using information we already know
# some sudoku can be solved without even needing to make a choice 
#(as you fill in certain answers, more certain answers will appear and eventually finish)

## coding wise the following
# initialization:
## run through the board the first time to put given values into the 3 logs (hori vert square)
## create any MIN items so that recursion dont have to recreate it each loop

#from my past experience with sudoku, there is never a scenario where it requires u to make 3 choices
# or u wont fill in all the blanks. Potentially it is because when 3 choices are avaliable for any 
# given box, it means mutiple final solutions

# part 1:
# we have to create a certain number filler
### this number filler will loop through the entire map each time, left to right, top to bottom
### on empty box, this will check all 3 current log to see if there are only 1 option to fill in
### then it will fill that number in and add it to the 3 logs
### then repeat and repeat until they find out that there are no box that would only have 1 option 
##### (needless to say it will loop through the entire map alot, sometimes only filling in a few boxes
##### in one loop)
# if it leads to a dead end where box cannot be filled, it returns false + fake board to inform the 
# previous that it is a dead end (fake board is only there to act as dummy output for uniform)
# if it completes, it returns true + correct board to tell the previous layers that it completed

# part 2:
# we will have to create a uncertainty choice maker, 
#which functions similar to above but targets 2 options
### this uncertainty will loop through the entire map each time as well top to bottom 
### on empty box, they will check all 3 logs whether there are only 2 options
### if there is, they will fill in the block, log it, and loop through the 2 options leading
### recursion to part 1
##### (this last part let part 1 check to see if the choice helped made any other boxes certain)
### it will recieve signal from next layer telling it if it completed or is a dead end
### if it is a dead end, it will reload the board,log, and other data and restart what it started 
### out with, but now with a new number since its a loop
##### (theoretically it should reach completion before/when the loop finishes)

# needless to say while it evaded alot of unneccessary trials, it ran alot of unnessary scanning
# it is hard to say which one is worse without testing 

# However, I implimented an addition to help it avoid running over boxes that were filled:
# idea is such as this:

### 1st- on the initualization we will run through the board the first time to find coordinates
##### of empty boxes and keep it in a data structure
### 2nd- then when doing the loops, we will only loop on boxes with coordinates in the structure
### 3-rd once we place a number in the box, we remove the coordinate (if its a choice, temp remove)

#this way when we loop through the boxes, we will only loop through empty one and reducing computation
#cost based on how many boxes were filled each recursion layer

### now question is: what data structure do we store the coordinates (x,y) in?
### do we store it in a dict? a list? create a linked list structure?
# this data structure needs to give constant time when removing items in the middle of the list
## since we dont know which node is a node with a certain answer/2 options
# it also needs to be iterable
##### I have tried 2 lists but removal requires O(n) time
##### linked list would not work as it will require removing coordiantes in
##### dictionary seems to be a possibility but if we store x in key and y in value, x cannot repeats
####### not to mention it will be hard to locate x key with y value

##### then i realized, i can use a string to make x unique, turn the keys into 'xy'
##### since i wont have a value now, i will not need a dictionary, instead i can use a set, which 
##### is also iterable

## using that, we can replace the double for loop system for coordinate and simply loop through the set
## removing coordinates in set when neccessary and translating string coordinate into int every time
## after some testing, the program is take to run only 1/2 to 3/4 time to run compared to former version
## however, it also made it more abstract

class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
################################################################################################
#part 2      
        #this recursion makes choices from 2, checks if incomplete or recurse to certainty # filler
        def uncertainRecursion(board,hori,vert,box,bRanges): 
            if len(bRanges[0])==0: #checks whether the coordinate set ran out
                return True,board # if it did, we filled in all the empty box successfully

            # this builds the coordinates from the coordinate set
            for x in bRanges[0].copy(): #must use .copy since bRange change with loop
                i=int(x[0])#hori index
                j=int(x[1]) #vert index
                k=((i//3)*3+1)+((j//3)%3)-1 #box index

                #unit set builder
                unitSet=set() #reset to make empty current set
                unitSet.update(bRanges[1]) #bRanges is just a collection set with 1-9
                for n in hori[i]: #based on hori vert and box(square), we remove the item from collection
                    unitSet.discard(n)
                for n in vert[j]:
                    unitSet.discard(n)
                for n in box[k]:
                    unitSet.discard(n) #the numbers that are left are valid options for filling the box
                if len(unitSet)==0: #if we find that there are no option for any box, its dead end
                    return False,board #return false and dummy wrong board
                elif len(unitSet)==2: ## when we have 2 options
                    tempBRanges=copy.deepcopy(bRanges) #first we make deep copy of each item
                    tempBoard=copy.deepcopy(board) #as they will be edited when it passes down
                    tempHori=copy.deepcopy(hori) # and we will need to reset if the path is dead
                    tempVert=copy.deepcopy(vert)
                    tempBox=copy.deepcopy(box)
                    for n in unitSet.copy(): #this might not need to be a soft copy but i did it anyways
                        bRanges[0].discard(f"{i}{j}") #update the coordinate set to remove filled item
                        board[i][j]=n #change the current board
                        hori[i].add(n)#horizontal logger to log new numb
                        vert[j].add(n)#vertical logger
                        box[k].add(n)#box logger
                        #then we enter the new certainty number filler
                        [complete,board]=setRecursion(board,hori,vert,box,bRanges) 
                        if complete: #if certainty number filler successfully filled up
                            return True,board # we pass good news and correct board
                        else: #else we need to reset all the items that could be changed by setRecurs
                            bRanges=tempBRanges # and try again next loop for next option
                            board=tempBoard
                            hori=tempHori
                            vert=tempVert
                            box=tempBox
################################################################################################
#part 1
        ## this is our certainty number filler, checks if complete or recurse to others
        def setRecursion(board,hori,vert,box,bRanges):
            setRecurse=False; # this tells us whether there are at least 1 certain options this run & 
            # whether we should do another certain run to check if filling those options create more
            # 1 options
            if len(bRanges[0])==0: #if we ran out of number from coordinate set, yay
                return True,board

            for x in bRanges[0].copy():#must use .copy since bRange change with loop, same as prev
                i=int(x[0])#hori index
                j=int(x[1]) #vert index
                k=((i//3)*3+1)+((j//3)%3)-1 #box index

                #unit set builder, same as previous
                unitSet=set()
                unitSet.update(bRanges[1])
                for n in hori[i]:
                    unitSet.discard(n)
                for n in vert[j]:
                    unitSet.discard(n)
                for n in box[k]:
                    unitSet.discard(n)
                if len(unitSet)==0: # same as previous, false checker if no options
                    return False,board
                elif len(unitSet)==1: #if only 1 option, we know this option must be it
                    bRanges[0].discard(f"{i}{j}") #safely discard the coordinate and mark as done
                    n=unitSet.pop() #extracting value from the option set
                    board[i][j]=n #add the value into board
                    hori[i].add(n)#horizontal logger
                    vert[j].add(n)#vertical logger
                    box[k].add(n)#box logger 
                    setRecurse=True #inform ourselves that we found a "1 option", and we should 
                    #redo to see if this 1 option we just filled in will make more "1 options"
            #if we filled in anything, we should check again to see if we can fill more
            if setRecurse: 
                [complete,board]=setRecursion(board,hori,vert,box,bRanges) #recieves news and final board
                return complete,board #pass news backwards 
             #if we loop through everything, there are still items in coordinate
            else: ## it means we have to make a choice on one of the empty box, or its a dead end
                [complete,board]=uncertainRecursion(board,hori,vert,box,bRanges) 
                #this will make choice or check if dead end
                return complete,board #pass news backwards

################################################################################################
#initialize              
        hori=[];vert=[];box=[]; #initalizing 9 by 9 set (9 sets in a list)
        bRanges=[]; #initualizing MIN items and coordinate set
        bLenRanges=range(len(board)) #MIN
        bRanges.append(set())#bRanges[0] is a coordinate set
        bRanges.append([]) #bRanges[1] is a MIN item for reset unitSet
        for i in range(1,len(board[0])+1):
            bRanges[1].append(f'{i}') #create MIN item for reset unitSet

        for i in bLenRanges: #create loggers
            hori.append(set())
            vert.append(set())
            box.append(set())
        #set initializer
        for i in bLenRanges: #hori index
            for j in bLenRanges: #vert index
                if board[i][j]!='.': #log everything already filled in in board
                    bIJ=board[i][j]
                    k=((i//3)*3+1)+((j//3)%3)-1 #i,j is the row and collumn index, k is box index
                    hori[i].add(bIJ)#horizontal logger
                    vert[j].add(bIJ)#vertical logger
                    box[k].add(bIJ)#box logger 
                else:
                    bRanges[0].add(f"{i}{j}") #for empty boxes, we add them into coordinates set
#set initializer
################################################################################################
        [complete,tboard]=setRecursion(board,hori,vert,box,bRanges) #start out with certainty filler
        #return tboard
        
### where it should end
################################################################################################
        ## which will recurse into uncertainty choice maker if needed
        ## normally i would just return board here...
        
        ## it took me 4 hrs to find out it reads the deep data for board rather than
        ## check what I gave it, which means it will always read the first solution (dead end or not)
        ## my board has
        
        # since it doesnt take the input i gave it, I must brute force change the input to my ans
        for i in bLenRanges:
            for j in bLenRanges:
                board[i][j]=tboard[i][j]
                
        return board #uneeded given how they check ans but included anyways
```

```python
#the idea was to create a recursion loop where it would count the current ans and 
#create an ans2 with count+ its counted numbers
#this is really just a reoccuring problem where result of one is input of next 
n=6;truAns="312211"
ni=n-2;ans='1' #ni(n counter) starts at 2 (hence n-2) because line we already handled when n=1 with ans='1'


def countAndLog(n,ans):
        if n<0:#kills recursion when counter exceeded
            return ans
        else:
            tempAns='' 
            count=1;temp=ans[0] #prepare logging when i=0 #count=1 to account 1st num
            #there are three parts: when i=0; when i=end;
            for i in range(len(ans)-1):  #When 0<i<end #this handles when 0<i<end # -1 to avoid out of bound
                ip=i+1 #MIN
                if ans[ip]==temp:  #scenario where next num==cur num (add count)
                    count+=1
                else:
                    tempAns+=str(count)+temp #scenario when it isnt (conclude/log count/solution, reset)
                    temp=ans[ip]
                    count=1 #count=1 to account 1st num
            tempAns+=str(count)+temp #this concludes when i=end (similar to else but no reset)
            ans=countAndLog(n-1,tempAns)
        return ans
k=countAndLog(ni,ans)
print(k)
```

```python
#someone else's slightly faster code, same proccess but looks cleaner and simpler
#instead of passing ans down like i did in the recursion and then back up,
#since we are building from grounds up with only int as inputs, he reverse the order by diving to the bottom first
#and work his way back up, evading the need of passing ans down
class Solution:
    def countAndSay(self, n: int) -> str:
        
        # Base case
        if n == 1:
            return "1"
        
        # Recursively get the string "saying" for n - 1
        saying = self.countAndSay(n - 1)
        
        # Ininialize output to empty string
        result = ""
        
        # Initialize counters
        last_char = saying[0]
        last_count = 0

        # Iterate though each character and compare to prior char
        for ch in saying:
            if ch == last_char:
                # Characters match, so increment count
                last_count += 1
            else:
                # Characters don't match, so append current counts to output and then reset counts
                result += str(last_count) + last_char
                last_char = ch
                last_count = 1

        # Append last counts to output and return
        result += str(last_count) + last_char
        return result          
```

```python
#simple recursion problem, branch out and get to all combination
#coding this in python presents a problem of deep copy
#hence I must carefully reset the permutation strings each time i add and remove a new number
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        
        def recursion(nums,permutation, ans):
            numsLen=len(nums) #MIN
            if numsLen==0: #record if reached the end of the depth
                ans.append(permutation.copy()) #must record copy since deep copy will directly edit ans
                return #return to skip next part but since numsLen is 0 rn, its not neccessary
            for i in range(numsLen):
                permutation.append(nums[i]) #add number
                recursion(nums[:i]+nums[i+1:], permutation, ans) #into the depth with number added
                permutation.pop() #remove number so it will be empty for next iter
                
            
            
        permutation=[];ans=[]
        recursion(nums,permutation, ans)
        return ans
```

```python
#O(n^2) dp using table without knowledge that neg previous sums are unhelpful
#location of max sum sublist naturally stored in the table
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        memo=[]
        ans=nums[0]
        for i in range(len(nums)):
            memo.append([]) #creating table
            for j in range(len(nums)-i):
                if j==0:
                    tempsum=nums[i] #base case
                else:
                    tempsum=memo[i][j-1]+nums[i+j] #adding current # with previous #
                memo[i].append(tempsum)
                if tempsum>ans:
                    ans=tempsum
        return ans
```

```python
#code from someone else that is about 16-20 ms faster
class Solution:
    def mySqrt(self, n: int) -> int:
        if n <= 1:
            return n
        
        start=0
        end=n//2
        ans=-1
        while start<=end:
            mid = start+(end-start)//2
            s=mid*mid
            if s==n:
                return mid
                break
            elif s<n:
                ans=mid
                start=mid+1
            else:
                end=mid-1
        return ans
        return ans
```
